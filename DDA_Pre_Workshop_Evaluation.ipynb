**Exercise 3.1**

Create the following 3 arrays.

ra  = range [10,50], 20 intervals <br>
dec = range [0,10] , 20 intervals <br>
mag = range [5,25] , 20 intervals <br>

```python
# your answer here in 3 lines
# code begins
ra  = np.linspace(10, 50, num=20)
dec = np.linspace(0, 10, num=20)
mag = np.linspace(5, 25, num=20)
# code ends
obj_type = ['star']*10 + ['galaxy']*10

df = pd.DataFrame(zip(ra, dec, mag, obj_type), 
                  columns = ['ra', 'dec', 'mag','obj_type'])
df
```


**Exercise 3.2**

Write code to select only the rows whose ra values are less than 35 and dec values greater than 4.

```python
# your answer here in 1 line
# code begins
# req_row=df[df[ra]<35] & df[df[dec]>4]
req_row=df[(df['ra']<35) &  (df['dec']>4.0)]
print(req_row)
# code ends
```

**Exercise 3.3**

Write code to create two new dataframes, df_star and df_galaxy, such that they are classified according to the object type. 

Hint: Use the *coloumn* obj_type

```python
# your answer here in 2 lines
# code begins
df_star=df[df['obj_type']=='star']

df_galaxy=df[df['obj_type']=='galaxy']

print(df_star)
print('\n')
print(df_galaxy)

# code ends
```

**Exercise 3.4**

Create scatter plot of ra vs dec for both dataframes, df_star and df_galaxy, with mag in color axis. Use cmap : 'Reds' and marker :'o' for stars and cmap 'Blues and marker '+' for galaxies. Show legends

```python
# your answer here in 3 lines
# code begins
plt.figure(figsize=(5,2))
plt.grid()
s=plt.scatter(x=df_star['ra'],y=df_star['dec'],marker='o',c=df_star['mag'],cmap="Reds")
plt.legend(["star"],loc="upper left")
plt.colorbar()
plt.show(s)
print('\n')

plt.figure(figsize=(5,2))
plt.grid()
g=plt.scatter(x=df_galaxy['ra'],y=df_galaxy['dec'],marker='+',c=df_galaxy['mag'],cmap="Blues")
plt.legend(["galaxy"],loc="upper left")
plt.colorbar()
plt.show(g)
print('\n')

plt.figure(figsize=(5,2))
plt.grid()
plt.scatter(x=df_star['ra'],y=df_star['dec'],marker='o',c=df_star['mag'],cmap="Reds")
plt.scatter(x=df_galaxy['ra'],y=df_galaxy['dec'],marker='+',c=df_galaxy['mag'],cmap="Blues")
plt.legend(["star","galaxy"],loc="upper left")
# code ends
```
